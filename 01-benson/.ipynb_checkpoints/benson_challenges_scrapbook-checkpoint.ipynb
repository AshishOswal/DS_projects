```python
# imports a library 'pandas', names it as 'pd'
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import dateutil.parser
import datetime as datetime


from IPython.display import Image

# enables inline plots, without it plots don't show up in the notebook
%matplotlib inline
```

```python
#Set display parameters
# set various options in pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows',150)
pd.set_option('display.precision', 3)
```

```python
df = pd.read_csv('data/turnstile_180310.csv')
df.shape
```

```python
df.columns = df.columns.str.strip()
```

```python
#Test the number of uniques of a given thing
#len(df.UNIT.value_counts())
len(df.STATION.value_counts())
```

```python
#create a unique string for each turnstile and add it as the 4th index
ts = df['C/A']+df.UNIT+df.SCP+df.STATION
df.insert(loc=4, column='TS', value=ts)
df
```

```python
#look at the number of unique turnstiles per station
gb=df.groupby(['STATION']).agg({'TS':lambda x: x.nunique()})
gb
```

```python
#create agg time column with datetime
agg_time = pd.to_datetime(df.DATE + " " + df.TIME)
df.insert(loc=7, column='agg_time', value=agg_time)
df
```

```python
#sort by TS and agg time for subtraction
df.sort_values(['TS', 'agg_time'])
```

```python
#create entries_diff just off of the sorted dataset
#df['entries_diff']= df['ENTRIES'].diff()
#df.head(100)
```

```python
#num < 0 when we just sort
#len(df[df['entries_diff']< 0])
```

```python
#number < 0 when we group by
len(df[df.groupby('TS')['ENTRIES'].diff()<0])
```

```python
#this doesn't make any difference
len(df[df.groupby('TS', sort='agg_time')['ENTRIES'].diff()<0])
```

```python
#Better way:
#create an entries_diff column that gives the difference between unit before for TS
#NANs will occur when it's the first entry for the turnstile
df['entries_diff']= df.groupby('TS')['ENTRIES'].diff()
#df.head(100)
```

```python
#number of entries_diff that are null
df['entries_diff'].isnull().sum()
```

```python
#number of entries_diff that are <0
len(df[df['entries_diff']< 0])
```

```python
#check for the number of entries_diff that are reaaly negative
df[df['entries_diff']< -10000].head(10)
```

```python
#Just look at wall street over time of day
gb=df[df.STATION == 'WALL ST'].groupby(['STATION','TIME']).agg({"entries_diff": np.sum,'TS':lambda x: x.nunique()})
gb

```

```python
#plot the station over time of day
plt.barh(gb.index.levels[1],gb.entries_diff)
```

```python
#get wall street for the dates
day=df[df.STATION == 'WALL ST'].groupby(['STATION','DATE']).agg({"entries_diff": np.sum,})
day
```

```python
plt.barh(day.index.levels[1],day.entries_diff)
```
