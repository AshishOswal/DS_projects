```python
#Challenge 1
def get_dictionary(df):
    d = {}
    for i, row in df.iterrows():
        key = (tuple(row[0:4]))
        value = list(row[4:])
        d[key] = d.get(key,[])+([value])
    return d
```

```python
def change_datetime_value(d):
    dt = {}
    for key, value in d.items():
        for i in value:
            newValue = []
            dayTime = i[2]+ ' '+i[3]
            datTimeFormatted = dateutil.parser.parse(dayTime)
            newValue.append(datTimeFormatted)
            newValue.append(i[5])
            dt[key] = dt.get(key,[])+([newValue])
    return dt

change_datetime_value(dictTiny)
```
