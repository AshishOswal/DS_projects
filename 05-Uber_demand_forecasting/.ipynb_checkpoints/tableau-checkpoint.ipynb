```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import numpy as np
import pandas as pd
import json
```

```python
neighborhoods = json.load(open("NY_neighborhoods.geojson.txt"))
rides_weather = pd.read_pickle("rides_weather.pkl")
```

```python
zones = rides_weather.columns[:-2]
```

```python
geo_names = []
for neighborhood in neighborhoods['features']:
    geo_names.append(neighborhood['properties']['neighborhood'])
```

```python
matches = dict()
for zone in zones:
    max_ratio = 0
    max_match = ""
    for geo in geo_names:
        if fuzz.ratio(zone,geo) > max_ratio:
            max_match = geo
            max_ratio = fuzz.ratio(zone,geo)
    matches[zone] = max_match
matches
```

```python
prediction = pd.read_pickle("prediction_df.pkl")
prediction.to_csv("prediction_df.csv")
```

```python
prediction = prediction.rename(columns=matches)
```

```python
tableau_df = pd.DataFrame(columns=['Datetime', 'Zone', 'Pred_Rides'])
for n in prediction.columns:
    temp = pd.DataFrame(columns=['Datetime', 'Zone', 'Pred_Rides'])
    temp["Datetime"] = prediction.index
    temp["Zone"] = n
    temp["Pred_Rides"] = prediction[n].values
    tableau_df = pd.concat([tableau_df,temp])
tableau_df = tableau_df.sort_values(["Datetime","Zone"])
tableau_df    
```

```python
tableau_agg = tableau_df.groupby(["Datetime","Zone"]).sum().reset_index()
```

```python
tableau_agg.to_csv("tableau_df_pred.csv")
```

```python
zones_ny = zones + ", NY"
def getzipcode(ser):
    zones_dict = dict()
    for zone in ser:
        try:
            latlng = gmaps.geocode(zone)[0]["geometry"]['location']
            zones_dict[zone[:-4]] = latlng
            if len(zones_dict) % 20 == 0:
                print("index =", len(zones_dict), "latlng =", latlng)
        except IndexError:
            print("index error at index=", len(zones_dict))
            pass
    return zones_dict
zone_dict = getzipcode(zones_ny)
```

```python
zone_df = pd.DataFrame(zone_dict)
```
