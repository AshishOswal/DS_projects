```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import numpy as np
import pandas as pd
import json
```

```python
neighborhoods = json.load(open("NY_neighborhoods.geojson.txt"))
rides_weather = pd.read_pickle("rides_weather.pkl")
```

```python
zones = rides_weather.columns[:-2]
```

```python
geo_names = []
for neighborhood in neighborhoods['features']:
    geo_names.append(neighborhood['properties']['neighborhood'])
```

```python
matches = dict()
for zone in zones:
    max_ratio = 0
    max_match = ""
    for geo in geo_names:
        if fuzz.ratio(zone,geo) > max_ratio:
            max_match = geo
            max_ratio = fuzz.ratio(zone,geo)
    matches[zone] = max_match
matches
```

```python
def create_tableau_df(filename):
    '''
    Takes in CSV file, pivots df to a form that Tableau prefers
    '''
    
    prediction = pd.read_csv(filename,index_col=0)
    prediction = prediction.rename(columns=matches)
    tableau_df = pd.DataFrame(columns=['Datetime', 'Zone', 'Pred_Rides'])
    for i, n in enumerate(prediction.columns):
        temp = pd.DataFrame(columns=['Datetime', 'Zone', 'Pred_Rides'])
        temp["Datetime"] = prediction.index
        temp["Zone"] = n
        temp["Pred_Rides"] = prediction.iloc[:,i].values
        tableau_df = pd.concat([tableau_df,temp])
    tableau_df = tableau_df.sort_values(["Datetime","Zone"])
    tableau_agg = tableau_df.groupby(["Datetime","Zone"]).sum().reset_index()
    return tableau_agg
```
